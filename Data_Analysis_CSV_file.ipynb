**Import Required Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: To make charts look better
sns.set(style="whitegrid")
```

**Load CSV File**

```python
# Example: Load file named 'sales_data.csv'
df = pd.read_csv("/content/sales_data.csv")

# Preview the data
df.head()
```

**Explore Dataset**

```python
# Get basic info
df.info()

# Get summary statistics
df.describe()

# Check for null values
df.isnull().sum()
```

**Perform Data Analysis**

*Total Sales by Productt*

```python
product_sales = df.groupby('Product_ID')['Sales_Amount'].sum().sort_values(ascending=False)
product_sales.plot(kind='bar', title='Total Sales by Product', figsize=(10,5))
plt.xlabel("Product")
plt.ylabel("Sales")
plt.tight_layout()
plt.show()
```

*Monthly Sales Trend (if date column exists)*

```python
df['Sale_Date'] = pd.to_datetime(df['Sale_Date'])  # Convert to datetime
df['Month'] = df['Sale_Date'].dt.to_period('M')

monthly_sales = df.groupby('Month')['Sales_Amount'].sum()
monthly_sales.plot(kind='line', marker='o', title='Monthly Sales Trend', figsize=(10,5))
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.tight_layout()
plt.show()
```

*Sales by Region/City *

```python
region_sales = df.groupby('Region_and_Sales_Rep')['Sales_Amount'].sum()
region_sales.plot(kind='pie', autopct='%1.1f%%', title='Sales by Region', figsize=(6,6))
plt.ylabel("")  # Hide y-label for pie chart
plt.show()
```
